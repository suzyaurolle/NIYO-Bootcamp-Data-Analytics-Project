replacing the null value with mean formula in our data set

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean',fill_value=None)
data['life_expectancy']=imputer.fit_transform(data[['life_expectancy']])
data['adult_mortality']=imputer.fit_transform(data[['adult_mortality']])
data['alcohol']=imputer.fit_transform(data[['alcohol']])
data['hepatitis_B']=imputer.fit_transform(data[['hepatitis_B']])
data['BMI']=imputer.fit_transform(data[['BMI']])
data['polio']=imputer.fit_transform(data[['polio']])
data['total_expenditure']=imputer.fit_transform(data[['total_expenditure']])
data['diphtheria']=imputer.fit_transform(data[['diphtheria']])
data['GDP']=imputer.fit_transform(data[['GDP']])
data['population']=imputer.fit_transform(data[['population']])
data['thinness_1_to_19_years']=imputer.fit_transform(data[['thinness_1_to_19_years']])
data['thinness_5_to_9_years']=imputer.fit_transform(data[['thinness_5_to_9_years']])
data['income_composition_of_resources']=imputer.fit_transform(data[['income_composition_of_resources']])
data['schooling']=imputer.fit_transform(data[['schooling']])
```

How are the variables in the data correlated with life expectancy

create a corrolation table of our data set

```python
data.corr().head()
```

```python
data.corr()['life_expectancy']
```

```python
corrmatrix = data.corr()
plt.subplots(figsize=(10, 10))
sns.heatmap(corrmatrix, square=True)
plt.title("Correlation Heatmap")
plt.show()
```

Which pairs of variables are highly correlated?

```python
corrmatrix_upper = corrmatrix.where(np.triu(np.ones(corrmatrix.shape), k=1).astype('bool'))
unique_pairs = corrmatrix_upper.unstack().dropna()
unique_pairs.abs()[unique_pairs.abs() > 0.8].round(3).sort_values(ascending=False)
```
